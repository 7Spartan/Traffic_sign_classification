```
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
from keras.layers import Dropout, Flatten0n
from keras.layers.convolutional import Conv2D, MaxPooling2D
import pickle
import pandas as pd
import random
```

```
np.random.seed(0)
```

```
with open('german-traffic-signs/train.p', 'rb') as f:
  train_data = pickle.load(f)
with open('german-traffic-signs/valid.p', 'rb') as f:
  val_data = pickle.load(f)
with open('german-traffic-signs/test.p', 'rb') as f:
  test_data = pickle.load(f)

print(type(train_data))

X_train, y_train = train_data['features'], train_data['labels']
X_val, y_val = val_data['features'], val_data['labels']
X_test, y_test = test_data['features'], test_data['labels']
```

```
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```
assert(X_train.shape[0]== y_train.shape[0]), "The number of images is not equal to the number of labels"
assert(X_val.shape[0]== y_val.shape[0]), "The number of val images is not equal to the number of labels"
assert(X_test.shape[0]== y_test.shape[0]), "The number of test images is not equal to the number of labels"
assert(X_train.shape [1:]==(32,32,3)), "The dimensions of the train images are not 32x32x3"
assert(X_val.shape [1:]==(32,32,3)), "The dimensions of the val images are not 32x32x3"
assert(X_test.shape [1:]==(32,32,3)), "The dimensions of the test images are not 32x32x3"
```

```
data = pd.read_csv('german-traffic-signs/signnames.csv')
num_of_samples=[]

col = 5 #To visualize the loaded images sample
num_classes = 43 #As tehre are 10 different digits to recognize

fig, axs = plt.subplots(nrows=num_classes, ncols=col, figsize =(5,50)) #subplots helps to plot multiple figures on same plot. Also returns tuples
fig.tight_layout()
for i in range(col):
    for j, row in data.iterrows():
        x_selected = X_train[y_train == j]
        axs[j][i].imshow(x_selected[random.randint(0,(len(x_selected)-1)),:,:],cmap=plt.get_cmap("gray"))
        axs[j][i].axis("off")
        if i ==2:
            axs[j][i].set_title(str(j) + '-' + row["SignName"])
            num_of_samples.append(len(x_selected))           
```

```
print(num_of_samples)
plt.figure(figsize=(12,4))
plt.bar(range(0,num_classes),num_of_samples)
plt.title("Distribution of the train data set")
plt.xlabel("Class number")
plt.ylabel("Number of images")
plt.show()
```

```
import cv2

plt.imshow(X_train[1000])
plt.axis("off")
print(X_train[1000].shape)
print(y_train[1000])
```

result shows (32,32,3) 3 as RGB image

```
def grayscale(img):
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  return img

img = grayscale(X_train[1000])
plt.imshow(img)
plt.axis("off")
print(img.shape)
```

```
#Histogram equalization aims to standardise lightining. As images are across several timings in day
#It would result in enhanced contrast too.
#It does so by normalising hte histogram of an image
```

```
def equalise(img):
  img = cv2.equalizeHist(img) #equalizeHist only takes in grayscale... it has no depth
  return img

img = equalise(img)
plt.imshow(img)
plt.axis("off")
print(img.shape)
```

```
def preprocessing(img):
  img = grayscale(img)
  img = equalise(img)
  img = img/255 #Normalises image pixels in between 0 and 1
  return img

X_train = np.array(list(map(preprocessing, X_train)))
X_val = np.array(list(map(preprocessing, X_val)))
X_test = np.array(list(map(preprocessing, X_test)))

```

```
plt.imshow(X_train[random.randint(0,len(X_train)-1)])
plt.axis("off")
print(X_train.shape)
```

```
X_train=X_train.reshape(34799,32,32,1) #See previous cell result for 34799
X_val=X_val.reshape(4410,32,32,1) #See previous cell result for 34799
X_test=X_test.reshape(12630,32,32,1) #See previous cell result for 34799
```

```
from keras.preprocessing.image import ImageDataGenerator #This class allows us to define the transformations on images
#indicate the max percentages of transformation that can be done
datagen = ImageDataGenerator(width_shift_range=0.1,
                             height_shift_range=0.1,
                             zoom_range=0.2,
                             shear_range=0.1,
                             rotation_range=10)
datagen.fit(X_train)
```

```
batches = datagen.flow(X_train, y_train, batch_size=20)
X_batch, y_batch = next(batches)

fig, axs = plt.subplots(1,15, figsize=(20,5))
fig.tight_layout()

for i in range(15):
  axs[i].imshow(X_batch[i].reshape(32,32))
  axs[i].axis("off")
```

```
print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```

```
y_train = to_categorical(y_train, 43)
y_val = to_categorical(y_val, 43)
y_test = to_categorical(y_test, 43)
```

```
def modified_model():
  model = Sequential()
  model.add(Conv2D(60, (5,5), input_shape=(32,32,1),activation='relu'))#60 filters or kernals, (5,5)size of filter, (32,32,1) is image size 
  model.add(Conv2D(60, (5,5), activation='relu')) #stride or padding arguments are not added and are left default in the above line
  model.add(MaxPooling2D(pool_size=(2,2))) #This will scale down by 2
  
  model.add(Conv2D(30,(3,3),activation = 'relu'))
  model.add(Conv2D(30,(3,3),activation = 'relu'))
  model.add(MaxPooling2D(pool_size = (2,2)))
#   model.add(Dropout(0.5))
  
  model.add(Flatten())
  model.add(Dense(500, activation = 'relu')) #500 is arbitrary adjusted as desired
  model.add(Dropout(0.5)) #50% nodes will be dropped at each update to prevent over fitting
  model.add(Dense(num_classes, activation='softmax'))
  model.compile(Adam(lr=0.001), loss ='categorical_crossentropy', metrics = ['accuracy'])
  return model
```

```
model=modified_model()
print(model.summary())
```

```
history = model.fit_generator(datagen.flow(X_train, y_train, batch_size=50),steps_per_epoch=2000, epochs = 10, validation_data=(X_val,y_val), shuffle =1)

# (X_train, y_train, epochs = 10, validation_data =(X_val,y_val),batch_size=400, verbose=1, shuffle=1)
```

```
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['training','validation'])
plt.title('Loss')
plt.xlabel('epoch')
```

```
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.legend(['training','validation'])
plt.title('Accuracy')
plt.xlabel('epoch')
```

```
score = model.evaluate(X_test,y_test, verbose=0)
print('Test Score:', score[0])
print('Test Accuracy:', score[1])
```
